```python
import os, shutil

# Paths
original_dataset_dir = r'D:\Data folder\dsplabs\ELECTRICAL ENGINEERING\SEMESTER 8th\Lab- machine learninng\ML_LAB\lab11\plantvillage dataset\color'
base_dir = r'D:\Data folder\dsplabs\ELECTRICAL ENGINEERING\SEMESTER 8th\Lab- machine learninng\ML_LAB\lab11\task1'

# Create base folders
if not os.path.exists(base_dir):
    os.mkdir(base_dir)

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

os.mkdir(train_dir)
os.mkdir(validation_dir)
os.mkdir(test_dir)

# Create subfolders
train_healthy_dir = os.path.join(train_dir, 'healthy')
train_unhealthy_dir = os.path.join(train_dir, 'unhealthy')
validation_healthy_dir = os.path.join(validation_dir, 'healthy')
validation_unhealthy_dir = os.path.join(validation_dir, 'unhealthy')
test_healthy_dir = os.path.join(test_dir, 'healthy')
test_unhealthy_dir = os.path.join(test_dir, 'unhealthy')

os.mkdir(train_healthy_dir)
os.mkdir(train_unhealthy_dir)
os.mkdir(validation_healthy_dir)
os.mkdir(validation_unhealthy_dir)
os.mkdir(test_healthy_dir)
os.mkdir(test_unhealthy_dir)

# Source directories (correct names)
healthy_src = os.path.join(original_dataset_dir, 'Corn_(maize)___healthy')
cercospora_src = os.path.join(original_dataset_dir, 'Corn_(maize)___Cercospora_leaf_spot Gray_leaf_spot')
common_rust_src = os.path.join(original_dataset_dir, 'Corn_(maize)__Common_rust_')
northern_leaf_blight_src = os.path.join(original_dataset_dir, 'Corn___(maize)_Northern_Leaf_Blight')

# Copy Healthy images
healthy_fnames = os.listdir(healthy_src)

for i, fname in enumerate(healthy_fnames[:120]):
    src = os.path.join(healthy_src, fname)
    dst = os.path.join(train_healthy_dir, f'healthy.{i}.jpg')
    shutil.copyfile(src, dst)

for i, fname in enumerate(healthy_fnames[120:150]):  # 30 for validation
    src = os.path.join(healthy_src, fname)
    dst = os.path.join(validation_healthy_dir, f'healthy.{i}.jpg')
    shutil.copyfile(src, dst)

for i, fname in enumerate(healthy_fnames[150:162]):  # 12 for testing
    src = os.path.join(healthy_src, fname)
    dst = os.path.join(test_healthy_dir, f'healthy.{i}.jpg')
    shutil.copyfile(src, dst)

# Function to copy unhealthy images
def copy_unhealthy_images(source_dir, prefix):
    filenames = os.listdir(source_dir)

    # Train: 40 images
    for i, fname in enumerate(filenames[:40]):
        src = os.path.join(source_dir, fname)
        dst = os.path.join(train_unhealthy_dir, f'{prefix}_train.{i}.jpg')
        shutil.copyfile(src, dst)
    
    # Validation: 10 images
    for i, fname in enumerate(filenames[40:50]):
        src = os.path.join(source_dir, fname)
        dst = os.path.join(validation_unhealthy_dir, f'{prefix}_val.{i}.jpg')
        shutil.copyfile(src, dst)

    # Test: 4 images
    for i, fname in enumerate(filenames[50:54]):
        src = os.path.join(source_dir, fname)
        dst = os.path.join(test_unhealthy_dir, f'{prefix}_test.{i}.jpg')
        shutil.copyfile(src, dst)

# Copy for each unhealthy class
copy_unhealthy_images(cercospora_src, 'cercospora')
copy_unhealthy_images(common_rust_src, 'common_rust')
copy_unhealthy_images(northern_leaf_blight_src, 'northern_leaf_blight')

print("✅ Dataset organized exactly as per Lab Task 1!")
```
